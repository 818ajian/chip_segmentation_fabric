```python
DEVICE = 0
USE_CUDA = torch.cuda.is_available()
def w(v):
    if USE_CUDA:
        return v.cuda(DEVICE)
    return v
```

```python
model = w(UNetClassifyS2(layers=4, init_filters=32, fusion_method='mul'))
weights = torch.load('../../weights/onera/unet_siamese_late_pooling_mul_relu_inp32_13band_2dates_focal_hm_cnc_all_14_cities.pt')
model.load_state_dict(weights)
```

```python
dataset = full_onera_loader('../../datasets/onera/', ['B01','B02', 'B03', 'B04','B05','B06','B07','B08','B8A','B09','B10','B11','B12'])
```

```python
cities = list(dataset.keys())
cities.sort()
```

```python
input_size = 32
for city in cities:
    img = dataset[city]['images']
    label = dataset[city]['labels']
    out = np.zeros((img.shape[2],img.shape[3]))
    
    for i in range(0,img.shape[2]//32):
        for j in range(0,img.shape[3]//32):
            if i*32 + input_size < img.shape[2] and j*32 + input_size < img.shape[3]:
                img1_cat1, img1_cat2, img1_cat3, img2_cat1, img2_cat2, img2_cat3, _ = onera_siamese_loader_late_pooling(dataset, city, i*32, j*32, 32)

                img1_cat1 = w(torch.from_numpy(np.asarray([img1_cat1])))
                img1_cat2 = w(torch.from_numpy(np.asarray([img1_cat2])))
                img1_cat3 = w(torch.from_numpy(np.asarray([img1_cat3])))

                img2_cat1 = w(torch.from_numpy(np.asarray([img2_cat1])))
                img2_cat2 = w(torch.from_numpy(np.asarray([img2_cat2])))
                img2_cat3 = w(torch.from_numpy(np.asarray([img2_cat3])))

                logits = model(img1_cat1, img1_cat2, img1_cat3, img2_cat1, img2_cat2, img2_cat3)
#                 print (logits.min(),logits.max())
                pred = F.sigmoid(logits) > 0.5
                pred = pred.data.cpu().numpy()[0]
                out[i*32:i*32+input_size,j*32:j*32+input_size] = pred    

                
    rgb1 = cv2.merge([img[1][0], img[2][0], img[3][0]])
    rgb2 = cv2.merge([img[1][1], img[2][1], img[3][1]])
    
    plt.imshow(out)
    plt.show()
    
    plt.imshow(label)
    plt.show()
#     f = plt.figure(figsize=(15,10))
    
    
#     f.add_subplot(1,2,1)
#     plt.imshow(out)
#     plt.axis('off')
#     plt.title('predicted')
    
#     f.add_subplot(1,2,2)
#     plt.imshow(rgb1)
#     plt.axis('off')
#     plt.title('rgb date1')
    
#     ####################
#     f.add_subplot(2,2,1)
#     plt.imshow(label)
#     plt.axis('off')
#     plt.title('label')
    
#     f.add_subplot(2,2,2)
#     plt.imshow(rgb2)
#     plt.axis('off')
#     plt.title('rgb date2')
    
#     plt.show(block=True)
    
```
