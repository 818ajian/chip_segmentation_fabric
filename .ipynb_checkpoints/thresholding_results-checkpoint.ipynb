```python
import os
import cv2
import glob

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
cities = os.listdir('../datasets/onera/train_labels/')
```

```python
for city in cities:
    gt = cv2.imread('../datasets/onera/train_labels/' + city + '/cm/new_' + city + '.png')
    pred = cv2.imread('../datasets/onera/results/arch32x64_hm_' + city + '.png')
    fig = plt.figure()
    ax1 = fig.add_subplot(1,2,1)
    ax1.imshow(gt)
    ax2 = fig.add_subplot(1,2,2)
    ret,thresh1 = cv2.threshold(pred,20,255,cv2.THRESH_BINARY)
    ax2.imshow(thresh1)
```
